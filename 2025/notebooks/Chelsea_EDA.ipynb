## IV. Merging Census Tract Data
### Converting Street Addresses
This is necessary to convert addresses to latitude/longitude using **Census Geocoder**, from which we can hopefully identify census tracts for each street address.  

```python
def latlongconvert(filename:str, destination:str):
    if filename == '57VOT_288799.txt':
        dataframe = df_dict[filename]
        addresses = pd.DataFrame()
        addresses["res_st_no"] = dataframe["Residential Address Street Number "].astype(str) + " " + dataframe["Residential Address Street Name "].str.title()
        addresses["city"] = "Chelsea"
        addresses["state"] = "MA"
        addresses["zip_code"] = "02150"
    else:
        dataframe = df_dict[filename]
        addresses = pd.DataFrame()
        addresses["res_st_no"] = dataframe["res_st_no"].astype(str) + " " + dataframe["st_nm"].str.title()
        addresses["city"] = "Chelsea"
        addresses["state"] = "MA"
        addresses["zip_code"] = "02150"

    # Reset the index and adjust it to start from 1
    addresses.reset_index(drop=True, inplace=True)
    addresses.index += 1

    # Write the DataFrame to a CSV file with the adjusted index
    addresses.to_csv(os.path.join(destination, filename[:-4] + ' addresses.csv'), header=False)

    #return addresses
```

```python
destination = '..\data\Chelsea_addresses\\'

for key in df_dict:
    latlongconvert(key, destination)
```

### Sort GeocodeResults files
By default, the outputs of Census Geocoder can be out of order (in terms of index).

```python
# Code generated by Copilot

# Define the path to your geocode files
path = r'..\data\Chelsea_Geocodes\*'

# Loop through each file in the directory
for file in glob.glob(path):
    try:
        # Read the file into a DataFrame
        df = pd.read_csv(file, header=None)
        
        # Sort the DataFrame by the first column (assuming the first column is index 0)
        df = df.sort_values(by=0, key=lambda col: col.astype(str).astype(int))
        
        # Write the sorted DataFrame back to a CSV file
        df.to_csv(file, index=False, header=False)
    except PermissionError as e:
        print(f"PermissionError: {e} - {file}")
    except Exception as e:
        print(f"Error: {e} - {file}")
```
