## II. Reading the data
Note: I am working on dealing with a few files that contain special characters; at least one file just had one such character and was relatively easy to fix, but there are other files that aren't as simple.

```python
#Adapted from code generated by GitHub Copilot
path = r'C:\Users\taiyo\OneDrive\Desktop\bu_files\2024-2025\DS594\data-viz-urban-league-voter\2025\data\Malden Election Data Part 1 (2020-2023)\*'
all_files = glob.glob(path)

absentee = []
vbm = []
voter = []

for filename in all_files:
    try:
        df = pd.read_csv(filename, delimiter='|', encoding='utf8')
    except UnicodeDecodeError:
        df = pd.read_csv(filename, delimiter='|', encoding='latin1')

    if filename.startswith('Absentee'):
        absentee.append(df)
    elif filename.startswith('VBM'):
        vbm.append(df)
    elif filename.startswith('Voter'):
        voter.append(df)
df_absentee = pd.concat(absentee, ignore_index=True)
df_vbm = pd.concat(vbm, ignore_index=True)
df_voter = pd.concat(voter, ignore_index=True)
```
