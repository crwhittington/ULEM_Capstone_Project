```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Chelsea_demographics_by_precinct.csv')
df.head()
```

```python
round(df.groupby(["ward_number", "precinct_number"])["median_household_income"].mean(), 2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Set style for better visualizations
plt.style.use('seaborn')
sns.set_palette("husl")

# Create a figure with multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Median Household Income Analysis by Ward and Precinct', fontsize=16)

# 1. Bar plot by ward and precinct
ward_precinct_income = df.groupby(['ward_number', 'precinct_number'])['median_household_income'].mean().unstack()
ward_precinct_income.plot(kind='bar', ax=axes[0,0])
axes[0,0].set_title('Median Household Income by Ward and Precinct')
axes[0,0].set_xlabel('Ward Number')
axes[0,0].set_ylabel('Median Household Income ($)')
axes[0,0].legend(title='Precinct Number')
axes[0,0].tick_params(axis='x', rotation=0)

# 2. Heatmap of income distribution
sns.heatmap(ward_precinct_income, annot=True, fmt='.0f', cmap='YlOrRd', ax=axes[0,1])
axes[0,1].set_title('Heatmap of Median Household Income')
axes[0,1].set_xlabel('Precinct Number')
axes[0,1].set_ylabel('Ward Number')

# 3. Box plot by ward
sns.boxplot(x='ward_number', y='median_household_income', data=df, ax=axes[1,0])
axes[1,0].set_title('Income Distribution by Ward')
axes[1,0].set_xlabel('Ward Number')
axes[1,0].set_ylabel('Median Household Income ($)')

# 4. Violin plot by precinct
sns.violinplot(x='precinct_number', y='median_household_income', data=df, ax=axes[1,1])
axes[1,1].set_title('Income Distribution by Precinct')
axes[1,1].set_xlabel('Precinct Number')
axes[1,1].set_ylabel('Median Household Income ($)')

plt.tight_layout()
plt.show()

# Additional visualization: Interactive plot using plotly
fig = px.scatter(df, 
                 x='ward_number', 
                 y='median_household_income',
                 color='precinct_number',
                 size='median_household_income',
                 hover_data=['ward_number', 'precinct_number', 'median_household_income'],
                 title='Interactive Scatter Plot of Median Household Income',
                 labels={'ward_number': 'Ward Number',
                        'median_household_income': 'Median Household Income ($)',
                        'precinct_number': 'Precinct Number'})
fig.show()
```
