```python
import pandas as pd
```

```python
df1 = pd.read_csv("pre_chelsea_2017.csv")
df2= pd.read_csv("pre_chelsea_2022.csv")
```

```python
print(df1.shape)
print(df2.shape)
```

```python
# Convert election date column to datetime for consistency
df1['date'] = pd.to_datetime(df1['date'])

# Add a marker column to track turnout (can be 1 for each vote)
df1['voted'] = 1

# Pivot to wide format: one row per voter, one column per election date
pivot_df = df1.pivot_table(
    index=['voter_id_number', 'last_name', 'first_name', 'ward_number', 'precinct_number', 'census_tract', 'date_of_birth', 'date_of_registration', 'latitude', 'longitude'],
    columns='date',
    values='voted',
    fill_value=0
).reset_index()

# Optional: rename the date columns to string format
pivot_df.columns = [col.strftime('%Y-%m-%d') if isinstance(col, pd.Timestamp) else col for col in pivot_df.columns]

pivot_df.head()
```
